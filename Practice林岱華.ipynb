```python
import pandas as pd
pd.set_option('display.max_colwidth', None)#setting the maximize string show
fb = pd.read_csv('nysu_10902_2019立委_research.csv')
politics = pd.read_csv('9th_legislator_promise.csv')
```

```python
liu = fb[fb.page_name == "林岱樺"]
liu.head(1)
```

```python
liu[liu.created_time_taipei == max(liu.created_time_taipei)]
```

```python
liu[liu.created_time_taipei == min(liu.created_time_taipei)]
```

```python
liu['new_date'] = pd.to_datetime(liu['created_time_taipei']).dt.date
liu['month_year'] = pd.to_datetime(liu['new_date']).dt.to_period('M')
liu.head(1)
```

```python
results = liu.groupby('month_year').sum()
results
```

```python
results.index.name = 'newhead'
results.reset_index(inplace=True)
results
```

```python
results['newhead'].map(str)
```

```python
import matplotlib.pyplot as plt

labels = results['newhead'].map(str)
width = 0.5

fig, ax = plt.subplots()
fig.set_size_inches(15, 8)

ax.bar(labels, results['like_count'], width, label='like_count')
ax.bar(labels, results['share_count'], width, label='share_count', bottom=results['like_count'])
ax.bar(labels, results['comment_count'], width, label='comment_cuont', bottom=results['share_count']+results['like_count'])

ax.legend()

plt.show()
```

為什麼2019年4月的討論數那麼多

```python
April = liu[liu.month_year == "2019-04"]
```

```python
results_april = April.groupby('new_date').sum()
results_april
```

```python
results_april.index.name = 'newhead'
results_april.reset_index(inplace=True)
results_april
```

reset_index 重製索引。inplace = True：不创建新的对象，直接对原始对象进行修改；

```python
labels = [str(i) for i in range(1,18)]

fig,ax= plt.subplots()
fig.set_size_inches(15, 8)

ax.plot(labels, results_april['like_count'], label='like_count')
ax.plot(labels, results_april['share_count'], label='share_count')
ax.plot(labels, results_april['comment_count'], label='comment_count')

ax.legend()

plt.show()
```
